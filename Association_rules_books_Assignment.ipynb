```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_items=apriori(books,min_support=0.2,use_colnames=True)
frequent_items
```

```python
rules=association_rules(frequent_items,metric='confidence',min_threshold=0.5)
rules
```

```python
#tweeking the values of support and confidence
frequent_items_1=apriori(books,min_support=0.1,use_colnames=True)
frequent_items_1
```

```python
rules_1=association_rules(df=frequent_items_1,metric='confidence',min_threshold=0.4)
rules_1
```

```python
frequent_items_2=apriori(df=books,min_support=0.09,use_colnames=True)
frequent_items_2
```

```python
rules_2=association_rules(df=frequent_items_2,min_threshold=0.6)
rules_2
```
